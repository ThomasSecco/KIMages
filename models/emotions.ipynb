```python
auth_key = "fc4d8fc2-4e49-44a3-acab-af4cd8013bb1:fx"  # Replace with your key
translator = deepl.Translator(auth_key)

x=[df['Photo motif'],df['Photo title'],df['Note']]
df=df.drop(columns=['Photo motif','Photo title','Note'])
```

```python
a=[]
for i in x[0]:
        translation = translator.translate_text(i, target_lang="EN-GB")
        a.append(translation.text)


b=[]
for i in x[1]:
        if i!='':
            translation = translator.translate_text(i, target_lang="EN-GB")
            b.append(translation.text)
        else:
            b.append(i)

c=[]
for i in x[2]:
        if i!='':
            translation = translator.translate_text(i, target_lang="EN-GB")
            c.append(translation.text)
        else:
            c.append(i)
```

```python
nlp = spacy.load("en_core_web_md")

def embedding(x):
    return nlp(x).vector

motif=list(map(embedding,a))
title=list(map(embedding,b))
note=list(map(embedding,c))
```

```python
df['Photo motif']=motif
df['Photo title']=title
df['Note']=note
```
